```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
import requests
import pandas as pd

url = 'https://new.aldi.us/products/dairy-eggs/eggs/k/75'
headers = {'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36"}
response = requests.get(url, headers=headers)
soup = soup(response.content, 'html.parser')
#print(soup.prettify())

eggs_text_elements = soup.find('div', class_= "product-grid")

eggs_list = []

for elem in eggs_text_elements:
    title = elem.find("div", class_="product-tile__name").text
    brand = elem.find("div", class_="product-tile__brandname").text
    price = elem.find("div", class_="base-price product-tile__price base-price--product-tile").text
    size = elem.find("div", class_="product-tile__unit-of-measurement").text

    eggs_each_product = {
        "Product": title,
        "Brand": brand,
        "Price": price,
        "Size": size
    }
    eggs_list.append(eggs_each_product)

#eggs_list
eggs_df = pd.DataFrame(eggs_list)
eggs_df["Size"] = eggs_df["Size"].str.replace(r'\s+', '', regex=True)
    # ^^this line removes all spaces in the size column
eggs_df["Price"] = eggs_df["Price"].str.replace(r'\n', '', regex=True)

eggs_df
```

```python
filtered_eggs_df = eggs_df[~eggs_df["Product"].str.contains("Liquid|Whites", regex=True)]
filtered_eggs_df["Ingredient"] = "Eggs"
filtered_eggs_df
```
